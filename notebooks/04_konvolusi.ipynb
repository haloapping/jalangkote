# Import Modules

```python
import matplotlib.pyplot as plt
import numpy as np

from scipy import ndimage
from skimage import io, util
```

# Konvolusi

Konvolusi merupakan proses menerapkan sebuah penapis (kernel atau mask) ke sebuah citra.

```python
lena_gray = io.imread("../test_images/lena_gray.bmp")
mask = np.array([[1, 2, 1],
                 [2, 4, 2],
                 [1, 2, 1]]) / 16
lena_gray_with_convolve = ndimage.convolve(input=lena_gray, weights=mask)

fig, ax = plt.subplots(nrows=1, ncols=2, tight_layout=True, figsize=(8, 8))

ax[0].set_title("Lena Gray")
ax[0].imshow(lena_gray, cmap=plt.cm.gray)
ax[0].axis("off")

ax[1].set_title("Lena Gray with Filter")
ax[1].imshow(lena_gray_with_convolve, cmap=plt.cm.gray)
ax[1].axis("off");
```
