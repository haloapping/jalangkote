# Import Modules

```python
import os
import matplotlib.pyplot as plt

from skimage import data, io
```

# Citra-Citra Populer pada Pengolahan Citra

```python
baboon = io.imread("../test_images/baboon_rgb.bmp")
cameraman = io.imread("../test_images/camera_gray.bmp")
girl_rgb = io.imread("../test_images/girl_rgb.bmp")
lena_rgb = io.imread("../test_images/lena_rgb.bmp")
peppers_rgb = io.imread("../test_images/peppers_rgb.bmp")

fig, ax = plt.subplots(nrows=1, ncols=5, tight_layout=True, figsize=(15, 15))

ax[0].imshow(baboon)
ax[0].axis("off")
ax[0].set_title("Baboon RGB")

ax[1].imshow(cameraman, cmap=plt.cm.gray)
ax[1].axis("off")
ax[1].set_title("Cameraman Gray")

ax[2].imshow(girl_rgb)
ax[2].axis("off")
ax[2].set_title("Girl RGB")

ax[3].imshow(lena_rgb)
ax[3].axis("off")
ax[3].set_title("Lena RGB")

ax[4].imshow(peppers_rgb)
ax[4].axis("off")
ax[4].set_title("Peppers RGB")

plt.show()
```
