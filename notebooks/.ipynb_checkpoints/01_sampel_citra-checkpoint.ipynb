```python
image_names = os.listdir("../test_images/")
images = [io.imread("../test_images/" + image_name) for image_name in image_names]

fig, ax = plt.subplots(nrows=3, ncols=6, tight_layout=True, figsize=(15, 15))

ax[0, 0].imshow(images[0])
ax[0, 0].axis("off")
ax[0, 0].set_title("Baboon RGB")

ax[0, 1].imshow(images[1], cmap=plt.cm.gray)
ax[0, 1].axis("off")
ax[0, 1].set_title("Barbara Gray")

ax[0, 2].imshow(images[2], cmap=plt.cm.gray)
ax[0, 2].axis("off")
ax[0, 2].set_title("Bird Gray")

ax[0, 3].imshow(images[3], cmap=plt.cm.gray)
ax[0, 3].axis("off")
ax[0, 3].set_title("Boat Gray")

ax[0, 4].imshow(images[4], cmap=plt.cm.gray)
ax[0, 4].axis("off")
ax[0, 4].set_title("Cameraman Gray")

ax[0, 5].imshow(images[5], cmap=plt.cm.gray)
ax[0, 5].axis("off")
ax[0, 5].set_title("Circles Gray")

ax[1, 0].imshow(images[6], cmap=plt.cm.gray)
ax[1, 0].axis("off")
ax[1, 0].set_title("Girl Gray")

ax[1, 1].imshow(images[7])
ax[1, 1].axis("off")
ax[1, 1].set_title("Girl RGB")

ax[1, 2].imshow(images[8], cmap=plt.cm.gray)
ax[1, 2].axis("off")
ax[1, 2].set_title("Goldhill Gray")

ax[1, 3].imshow(images[9], cmap=plt.cm.gray)
ax[1, 3].axis("off")
ax[1, 3].set_title("Lena Gray")

ax[1, 4].imshow(images[10])
ax[1, 4].axis("off")
ax[1, 4].set_title("Lena RGB")

ax[1, 5].imshow(images[11], cmap=plt.cm.gray)
ax[1, 5].axis("off")
ax[1, 5].set_title("Mountain Gray")

ax[2, 0].imshow(images[12], cmap=plt.cm.gray)
ax[2, 0].axis("off")
ax[2, 0].set_title("Peppers Gray")

ax[2, 1].imshow(images[13])
ax[2, 1].axis("off")
ax[2, 1].set_title("Peppers RGB")

ax[2, 2].imshow(images[14], cmap=plt.cm.gray)
ax[2, 2].axis("off")
ax[2, 2].set_title("San Fransisco Gray")

ax[2, 3].imshow(images[15], cmap=plt.cm.gray)
ax[2, 3].axis("off")
ax[2, 3].set_title("Slope Gray")

ax[2, 4].imshow(images[16], cmap=plt.cm.gray)
ax[2, 4].axis("off")
ax[2, 4].set_title("Text Gray")

ax[2, 5].imshow(images[17], cmap=plt.cm.gray)
ax[2, 5].axis("off")
ax[2, 5].set_title("Zelda Gray")

plt.show()
```
