# Import Modules

```python
import matplotlib.pyplot as plt
import numpy as np

from scipy import ndimage, signal
from skimage import io, util
```

# Perbaikan Citra menggunakan Konvolusi

## Penapis Linier (Penapis Lanjar)

Low Pass Filter

### Pelembutan Citra (Image Smoothing)

```python
lena_with_noise = plt.imread("../test_images/lena_with_salt_pepper_noise.bmp")
size = 3
mask = np.ones((size, size)) / (size * size)
lena_with_smooth_filter = ndimage.convolve(lena_with_noise, mask)

fig, ax = plt.subplots(nrows=1, ncols=2, tight_layout=True, figsize=(7, 7))

ax[0].set_title("Image with Noise")
ax[0].axis("off")
ax[0].imshow(lena_with_noise, plt.cm.gray)

ax[1].set_title("Image with Smooth Filter")
ax[1].axis("off")
ax[1].imshow(lena_with_smooth_filter, plt.cm.gray);
```

### Pelembutan Gaussian (Gaussian Smoothing)

```python
lena_with_noise = plt.imread("../test_images/lena_with_salt_pepper_noise.bmp")
mask = np.array([[1, 2, 1],
                 [2, 4, 2],
                 [1, 2, 1]]) / 16

lena_with_smooth_filter = ndimage.convolve(lena_with_noise, mask)

fig, ax = plt.subplots(nrows=1, ncols=2, tight_layout=True, figsize=(7, 7))

ax[0].set_title("Image with Noise")
ax[0].axis("off")
ax[0].imshow(lena_with_noise, plt.cm.gray)

ax[1].set_title("Image with Gaussian Filter")
ax[1].axis("off")
ax[1].imshow(lena_with_smooth_filter, plt.cm.gray);
```

## Penapis  Non Linier (Nirlanjar)

### Penapis Minimum

Nilai piksel digantikan dengan nilai median dari sekumpulan nilai pada penapis.

```python
lena_with_noise = plt.imread("../test_images/lena_with_salt_noise.bmp")
lena_with_min_filter = ndimage.minimum_filter(lena_with_noise, size=(2, 2))

fig, ax = plt.subplots(nrows=1, ncols=2, tight_layout=True, figsize=(7, 7))

ax[0].set_title("Image with Noise")
ax[0].axis("off")
ax[0].imshow(lena_with_noise, plt.cm.gray)

ax[1].set_title("Image with Minimum Filter")
ax[1].axis("off")
ax[1].imshow(lena_with_min_filter, plt.cm.gray);
```

### Penapis Maksimum

```python
lena_with_noise = plt.imread("../test_images/lena_with_pepper_noise.bmp")
lena_with_max_filter = ndimage.maximum_filter(lena_with_noise, size=(2, 2))

fig, ax = plt.subplots(nrows=1, ncols=2, tight_layout=True, figsize=(7, 7))

ax[0].set_title("Image with Noise")
ax[0].axis("off")
ax[0].imshow(lena_with_noise, plt.cm.gray)

ax[1].set_title("Image with Minimum Filter")
ax[1].axis("off")
ax[1].imshow(lena_with_max_filter, plt.cm.gray);
```

### Penapisan Median (Nilai Tengah)

```python
lena_with_noise = plt.imread("../test_images/lena_with_salt_pepper_noise.bmp")
lena_with_med_filter = ndimage.median_filter(lena_with_noise, size=(2, 2))

fig, ax = plt.subplots(nrows=1, ncols=2, tight_layout=True, figsize=(7, 7))

ax[0].set_title("Image with Noise")
ax[0].axis("off")
ax[0].imshow(lena_with_noise, plt.cm.gray)

ax[1].set_title("Image with Minimum Filter")
ax[1].axis("off")
ax[1].imshow(lena_with_med_filter, plt.cm.gray);
```

### Penajaman Citra (Image Sharpening)

```python
lena_rgb = plt.imread("../test_images/lena_rgb.bmp")
red_channel_lena = lena_rgb[:, :, 0]
green_channel_lena = lena_rgb[:, :, 1]
blue_channel_lena = lena_rgb[:, :, 2]
lena_with_sharp_filter = np.zeros((lena_rgb.shape[0], lena_rgb.shape[1], 3), dtype=np.uint8)

mask = np.array([[0,  1, 0],
                 [1, -4, 1],
                 [0,  1, 0]])

red_channel_lena_with_sharp_filter = ndimage.convolve(red_channel_lena, mask)
green_channel_lena_with_sharp_filter = ndimage.convolve(green_channel_lena, mask)
blue_channel_lena_with_sharp_filter = ndimage.convolve(blue_channel_lena, mask)
lena_with_sharp_filter[:, :, 0] = red_channel_lena_with_sharp_filter
lena_with_sharp_filter[:, :, 1] = green_channel_lena_with_sharp_filter
lena_with_sharp_filter[:, :, 2] = blue_channel_lena_with_sharp_filter

fig, ax = plt.subplots(nrows=1, ncols=2, tight_layout=True, figsize=(7, 7))

ax[0].set_title("Original Image")
ax[0].axis("off")
ax[0].imshow(lena_rgb)

ax[1].set_title("Image with Sharp Filter")
ax[1].axis("off")
ax[1].imshow(lena_with_sharp_filter);
```

### Unsharp Masking

```python
lena_rgb = plt.imread("../test_images/lena_rgb.bmp")
red_channel_lena = lena_rgb[:, :, 0]
green_channel_lena = lena_rgb[:, :, 1]
blue_channel_lena = lena_rgb[:, :, 2]
lowpass = np.zeros((lena_rgb.shape[0], lena_rgb.shape[1], 3), dtype=np.uint8)

mask = np.ones((5, 5)) / 25

red_channel_lena_with_sharp_filter = ndimage.convolve(red_channel_lena, mask)
green_channel_lena_with_sharp_filter = ndimage.convolve(green_channel_lena, mask)
blue_channel_lena_with_sharp_filter = ndimage.convolve(blue_channel_lena, mask)
lowpass[:, :, 0] = red_channel_lena_with_sharp_filter
lowpass[:, :, 1] = green_channel_lena_with_sharp_filter
lowpass[:, :, 2] = blue_channel_lena_with_sharp_filter
highpass = lena_rgb - lowpass
lena_with_sharp_filter = lena_rgb + highpass
fig, ax = plt.subplots(nrows=2, ncols=2, tight_layout=True, figsize=(7, 7))

ax[0][0].set_title("Original Image")
ax[0][0].axis("off")
ax[0][0].imshow(lena_rgb)

ax[0][1].set_title("Highpass Image")
ax[0][1].axis("off")
ax[0][1].imshow(highpass)

ax[1][0].set_title("Lowpass Image")
ax[1][0].axis("off")
ax[1][0].imshow(lowpass)

ax[1][1].set_title("Sharp Image")
ax[1][1].axis("off")
ax[1][1].imshow(lena_with_sharp_filter);
```

### High Boost Filter

```python
lena_gray = plt.imread("../test_images/lena_gray.bmp")
mask = np.ones((5, 5)) / 25
lowpass = ndimage.convolve(lena_gray, mask)
highpass = lena_gray - lowpass
alpha = 2.4
lena_with_high_boost_filter = ((alpha - 1) * lena_gray) + highpass

fig, ax = plt.subplots(nrows=1, ncols=3, tight_layout=True, figsize=(10, 10))

ax[0].set_title("Original Image")
ax[0].axis("off")
ax[0].imshow(lena_gray, plt.cm.gray)

ax[1].set_title("Highpass Image")
ax[1].axis("off")
ax[1].imshow(highpass, plt.cm.gray)

ax[2].set_title("Highboost Image")
ax[2].axis("off")
ax[2].imshow(lena_with_high_boost_filter, plt.cm.gray);
```
